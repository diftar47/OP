## Завдання 1
**Перевірте рекурсивний алгоритм знаходження факторіалу**


```
def Fact(n)
{
    if (n==0)
        return 1; // умова завершення рекурсивного процеса
    else
        return n*Fact(n-1); // перехід до наступного числа n-1
}
```










**Завдання для самостійної роботи:**

Реалізуйте алгоритм знаходження факторіалу з вводом числа n від користувача:



````python
import time

def Fact(n):
    if n == 0:
        return 1
    else:
        return n * Fact(n-1)

x = int(input("Введіть n: "))

start_time = time.perf_counter()  # Начало замера времени
result = Fact(x)
end_time = time.perf_counter()    # Конец замера времени

print(f"Результат: {result}")
print(f"Час виконання: {end_time - start_time:.6f} секунд")


#Введіть n: 100
#Результат: 933262154...
#Час виконання: 0.000019 секунд
````

## Завдання 2
Напишіть НЕ рекурсивний алгоритм знаходження факторіалу числа n


````python
import time

def factorial(n):
    if n < 0:
        raise ValueError("Факторіал визначений тільки для невід'ємних чисел")
    if n == 0 or n == 1:
        return 1

    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

x = int(input("Введіть n: "))

start_time = time.perf_counter()  # Начало замера времени
result = factorial(x)
end_time = time.perf_counter()    # Конец замера времени

print(f"Результат: {result}")
print(f"Час виконання: {end_time - start_time:.6f} секунд")

#Введіть n: 100
#Результат: 933262154...
#Час виконання: 0.000017 секунд
````

10.Підрахунок кількості голосних у рядку
Рекурсивна функція, яка проходить рядок і рахує кількість голосних літер.

````python
def gcd(a, b):
    # Базовий випадок: якщо b дорівнює 0, повертаємо a
    if b == 0:
        return a
    # Рекурсивний виклик з новими аргументами (b, a % b)
    return gcd(b, a % b)

# Приклад використання
a = 48
b = 18
result = gcd(a, b)
print(f"НСД({a}, {b}) = {result}")  # Виведе: НСД(48, 18) = 6
````

## Контрольні запитання
1.Що таке рекурсія і які два основні елементи має рекурсивна функція?


Рекурсія - це техніка, коли функція викликає саму себе. Два основні елементи:
- Базовий випадок - умова виходу з рекурсії
- Рекурсивний випадок - виклик тієї ж функції з іншими параметрами

2.Як працює стек викликів при рекурсивному виконанні функції?
    
    
- Кожен рекурсивний виклик додає новий кадр у стек викликів. Коли досягається базовий випадок, стек починає "розгортатись" - кадри послідовно видаляються, поки не повернемося до початкового виклику.

3.Які потенційні проблеми можуть виникнути при використанні рекурсії?

- Переповнення стека (Stack Overflow) при глибокій рекурсії

- Вища витрата пам'яті порівняно з ітерацією

- Можлива менша ефективність через накладні витрати на виклики функцій
